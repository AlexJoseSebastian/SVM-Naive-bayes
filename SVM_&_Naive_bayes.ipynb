1: What is a Support Vector Machine (SVM)?
- Ans) SVM is a supervised learning algorithm used for classification and regression.

- It finds the optimal hyperplane that:
- Separates classes


2: What is the difference between Hard Margin and Soft Margin SVM?
- Ans)
- Hard Margin
  - Data must be perfectly separable.
  - Sensitive to outliers.
  - Used when data is clean.
- Soft Margin
  - Allowed some misclassification
  - Robust to noise
  - Used in real-world noisy data.

3: What is the mathematical intuition behind SVM?
- Ans) It's find the optimal hyperplane that separates classes by maximizing the margin—the distance between the hyperplane and the closest data points, known as "support vectors". It solves a constrained optimization problem to define a "fat" boundary, enhancing generalization to unseen data

4: What is the role of Lagrange Multipliers in SVM?
- Ans) Lagrange multipliers are crucial in SVM for converting the hard-to-solve primal optimization problem into a dual, convex quadratic programming problem that is easier to solve.

5: What are Support Vectors in SVM?
- Ans)
- Point closest to the decision boundary.
- Points defining the margin
- Points with non-zero Langrange multipliers.
6: What is a Support Vector Classifier (SVC)?
- Ans) A Support Vector Classifier (SVC) is a supervised machine learning algorithm that classifies data by finding the optimal hyperplane—a decision boundary—that maximizes the margin (distance) between different classes .
7: What is a Support Vector Regressor (SVR)?
- Ans) SVR applies SVM for regression.
- Fits a funtion.
- Allows error within epsilon.
- Ignores small deviations.

8: What is the Kernel Trick in SVM?
- Ans) It maps data to higher dimension wihtout explicitly computing transformation.

9: Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.
- Ans)
  - Linear being fastest for simple, high-dimensional data.
  - Polynomial capturing moderate, curved relationships.
  - RBF (Radial Basis Function) providing the most flexibility for complex, non-linear, or unknown patterns.

10: What is the effect of the C parameter in SVM?
- Ans)
- C controls trade-off between:
  - Margin width
  - Classification error

- Large C:
  - Small margin
  - Less misclassification
  - Risk of overfitting

- Small C:
  - Large margin
  - More tolerance
  - Risk of underfitting

11: What is the role of the Gamma parameter in RBF Kernel SVM?
- Ans) Gamma controls infulence of each training point.
- High gamma:
  - complex decision boundary
  - Overfitting risk
- Low gamma:
  - Smooth boundary.
  - Underfitting risk

12: What is the Naïve Bayes classifier, and why is it called "Naïve"?
- Ans) Naive Bayes is a fast, efficient supervised machine learning classification algorithm based on Bayes' Theorem, commonly used for text classification, spam filtering, and sentiment analysis.

13: What is Bayes' Theorem?
- Ans) Bayes' Theorem is a fundamental formula in probability theory that calculates the likelihood of an event based on prior knowledge of conditions related to it. It updates existing beliefs (prior probability) with new evidence (likelihood) to produce a revised, more accurate probability (posterior probability), commonly used in medical testing, spam filtering.
- Posterior = Likelihood × Prior / Evidence.

14: Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes?
- Ans)
- Gaussian Naive Bayes
  - Continuous Data Type .
  - Features follow a normal (Gaussian) distribution (bell-shaped curve).
  - Ideal for data with numerical, continuous variables, such as medical data or sensor data.
  - Calculates the mean and standard deviation of each feature per class.

- Multinomial Naïve Bayes
 - Discrete, non-negative integer counts Data Type.
 - Features follow a multinomial distribution, representing the frequency of occurrences.
 - Highly efficient for text classification where term frequencies (word counts) are used.
 - Models the probability of a word appearing in a document given the class.

- Bernoulli Naïve Bayes
 - Binary (Boolean) features.
 - Features are boolean, indicating presence (1) or absence (0).
 - Good for text classification based on whether a word exists in a document, rather than how often it appears (e.g., spam detection).
 - Penalizes the non-occurrence of a feature, which Multinomial ignores.

15: When should you use Gaussian Naïve Bayes over other variants?
- Ans)
- Features are continuous.
- Data roughly normally distributed.
- Small dataset.
- Need fast baseline.

16: What are the key assumptions made by Naïve Bayes?
- Ans)
- Conditional independence
- Correct feature distribution assumption
- Features contribute independently.

17: What are the advantages and disadvantages of Naïve Bayes?

- Advantages
 - Very fast
 - Works well with high-dimensional data
 - Good for text classification
 - Requires small training data
- Disadvantages
 - Unrealistic independence assumption
 - Poor with correlated features.
 - Probability estimates may not be accurate.

18: Why is Naïve Bayes a good choice for text classification?
- Ans)
- Text data is high-dimensional
- Words treated independently
- Works well with bag-of-words
- Fast training and prediction
- Used in:
  - spam filtering, Sentimental analysis.

19: Compare SVM and Naïve Bayes for classification tasks?
- Ans)
- SVM is generally more accurate for complex, high-dimensional data (text, images) by finding optimal margins SVM excels in accuracy but requires more computational resources.

- Naive Bayes is faster, simpler, and efficient for smaller datasets or when features are independent.

20: How does Laplace Smoothing help in Naïve Bayes?
- Ans) Laplace smoothing (or additive smoothing) prevents Naive Bayes classifiers from failing when a feature (e.g., a word) in the testing data never appeared in the training set, which would otherwise result in a zero probability.


21) Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy?

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

model = SVC(kernel='linear')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

22) Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies?

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

wine = datasets.load_wine()
X = wine.data
y = wine.target

scaler = StandardScaler()
X = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

linear_svm = SVC(kernel='linear')
linear_svm.fit(X_train, y_train)
linear_pred = linear_svm.predict(X_test)
linear_acc = accuracy_score(y_test, linear_pred)

rbf_svm = SVC(kernel='rbf')
rbf_svm.fit(X_train, y_train)
rbf_pred = rbf_svm.predict(X_test)
rbf_acc = accuracy_score(y_test, rbf_pred)

print("Linear Kernel Accuracy:", linear_acc)
print("RBF Kernel Accuracy:", rbf_acc)
```

24) Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary ?


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

X, y = make_blobs(n_samples=100, centers=2, random_state=42)

model = SVC(kernel='poly', degree=3)
model.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx, yy = np.meshgrid(
    np.linspace(xlim[0], xlim[1], 100),
    np.linspace(ylim[0], ylim[1], 100)
)

Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contour(xx, yy, Z, levels=[0], linewidths=2)
plt.title("Polynomial Kernel SVM Decision Boundary")
plt.show()
```

27 Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually .


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

X, y = make_blobs(n_samples=200, centers=2, random_state=42)

C_values = [0.1, 1, 100]

plt.figure(figsize=(12,4))
for i, C in enumerate(C_values):
    model = SVC(kernel='linear', C=C)
    model.fit(X, y)

    plt.subplot(1,3,i+1)
    plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm')

    ax = plt.gca()
    xlim, ylim = ax.get_xlim(), ax.get_ylim()
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1],100),
                         np.linspace(ylim[0], ylim[1],100))
    Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contour(xx, yy, Z, levels=[0])
    plt.title(f"C={C}")

plt.show()
```

29 Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data?

```python
from sklearn.datasets import load_wine
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_wine()
X_train,X_test,y_train,y_test = train_test_split(data.data,data.target,test_size=0.3,random_state=42)

# Without scaling
model1 = SVC()
model1.fit(X_train,y_train)
acc1 = accuracy_score(y_test, model1.predict(X_test))

# With scaling
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

model2 = SVC()
model2.fit(X_train_s,y_train)
acc2 = accuracy_score(y_test, model2.predict(X_test_s))

print("Unscaled Accuracy:", acc1)
print("Scaled Accuracy:", acc2)
```

30 Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before and after Laplace Smoothing ?


```python
from sklearn.datasets import load_breast_cancer
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(data.data,data.target,test_size=0.3,random_state=42)

# Default
model1 = GaussianNB()
model1.fit(X_train,y_train)
acc1 = accuracy_score(y_test, model1.predict(X_test))

# With smoothing (var_smoothing parameter)
model2 = GaussianNB(var_smoothing=1e-2)
model2.fit(X_train,y_train)
acc2 = accuracy_score(y_test, model2.predict(X_test))

print("Default Accuracy:", acc1)
print("With Smoothing Accuracy:", acc2)
```

32 Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and check it improve accuracy ?


```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X,y = make_classification(n_samples=1000, weights=[0.9,0.1], random_state=42)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

# Without weighting
model1 = SVC()
model1.fit(X_train,y_train)
acc1 = accuracy_score(y_test, model1.predict(X_test))

# With class weighting
model2 = SVC(class_weight='balanced')
model2.fit(X_train,y_train)
acc2 = accuracy_score(y_test, model2.predict(X_test))

print("Without Weight:", acc1)
print("With Weight:", acc2)
```

35 Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare results ?


```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(data.data,data.target,test_size=0.3)

selector = SelectKBest(chi2, k=10)
X_train_sel = selector.fit_transform(X_train,y_train)
X_test_sel = selector.transform(X_test)

model = GaussianNB()
model.fit(X_train_sel,y_train)
print("Accuracy after selection:", accuracy_score(y_test, model.predict(X_test_sel)))
```

36 Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and One-vs-One (OvO) strategies on the Wine dataset and compare their accuracy ?

40 Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy?


```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(data.data,data.target,test_size=0.3)

model = SVC(kernel='linear')
rfe = RFE(model, n_features_to_select=10)
X_train_rfe = rfe.fit_transform(X_train,y_train)
X_test_rfe = rfe.transform(X_test)

model.fit(X_train_rfe,y_train)
print("Accuracy after RFE:", accuracy_score(y_test, model.predict(X_test_rfe)))
```

46 Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

data = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(data.data,data.target,test_size=0.3)

model = SVC(probability=True)
model.fit(X_train,y_train)

probs = model.predict_proba(X_test)[:,1]
precision, recall, _ = precision_recall_curve(y_test, probs)

plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.show()
```
